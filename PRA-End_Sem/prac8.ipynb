```python
from hmmlearn import hmm
import numpy as np
```

```python
symbol_map = {'A' : 0, 'G' : 1, 'T' : 2, 'C' : 3}
sequence = 'ACGTCGACGGTACGT'
```

```python
X = np.array([[symbol_map[nuc] for nuc in sequence]])
length = len(X)
```

```python
model = hmm.MultinomialHMM(n_components = 2, n_iter = 100,init_params = 'ste', params = 'ste', random_state = 42)
model._init(X,length)
model.fit(X,length)
```

```python
# Predict the hidden states
hidden_states = model.predict(X)

# Output
print("DNA Sequence:", sequence)
print("Predicted Hidden States (0 or 1):", hidden_states)
```
